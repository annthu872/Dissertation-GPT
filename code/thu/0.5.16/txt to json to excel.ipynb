```python
import os

def create_summary_files(directory, path):
    # Define the names of the summary files
    directory_name = os.path.basename(directory)
    summary_file_raw = os.path.join(path, f"{directory_name}-Raw.txt")
    summary_file = os.path.join(path, f"{directory_name}.txt")

    # Initialize strings to store file contents
    raw_content = ""
    non_raw_content = ""

    # Iterate through the files in the directory
    for file_name in os.listdir(directory):
        file_path = os.path.join(directory, file_name)
        # Check if the current item is a file and not one of the summary files
        if os.path.isfile(file_path) and file_name not in [os.path.basename(summary_file_raw), os.path.basename(summary_file)]:
            # Read the file content
            with open(file_path, 'r') as file:
                content = file.read()
                print(content)
                if "-Raw" in file_name:
                    raw_content += "Scenario: "+ file_name +"\n"
                    raw_content += content + "\n"
                else:
                    non_raw_content +="Scenario: "+ file_name +"\n"
                    non_raw_content += content + "\n"

    # Write the collected raw file contents to "Summary File Raw.txt"
    with open(summary_file_raw, 'w') as file:
        file.write(raw_content)

    # Write the collected non-raw file contents to "Summary File.txt"
    with open(summary_file, 'w') as file:
        file.write(non_raw_content)

# Example usage
directory_path =r"D:\Dissertation-GPT\5.25-NhuY\Matcha\AddLesson"  # Replace with your folder path
path = r"D:\Dissertation-GPT\5.25-NhuY\Matcha\summary test case"
create_summary_files(directory_path, path)
```

```python
def get_test_steps_and_expected_result(scenario, file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
        
        # Split content into individual scenarios
        scenarios = content.split("Scenario:")
        scenarios = content.strip().split('Scenario: ')
        scenarios = [s for s in scenarios if s.strip()]  # Remove empty strings if any
        for scenario_block in scenarios:
            if scenario in scenario_block:
                lines = scenario_block.strip().split("\n")
                test_steps = ""
                expected_result = ""
                for line in lines:
                    if line.startswith("testSteps:"):
                        test_steps = line.split("testSteps:")[1].strip()
                    elif line.startswith("expectedResult:"):
                        expected_result = line.split("expectedResult:")[1].strip()
                return test_steps, expected_result

        return None, None  # Scenario not found
    except FileNotFoundError:
        print("File not found.")
        return None, None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None, None

# Example usage:
scenario = "Cancel Adding New Lesson After Entering Name1"  # Replace with the actual scenario you are looking for
file_path = r"D:\Dissertation-GPT\5.25-NhuY\Matcha\summary test case\AddLesson-Raw.txt"  # Replace with the actual file path
test_steps, expected_result = get_test_steps_and_expected_result(scenario, file_path)

if test_steps and expected_result:
    print("Test Steps:", test_steps)
    print("Expected Result:", expected_result)
else:
    print("Scenario not found or an error occurred.")
```

```python
import json

content = ""
with open(r"D:\Dissertation-GPT\5.25-NhuY\Matcha\summary test case\AddLesson-Raw.txt", 'r') as file:
    content = file.read()
scenarios = content.strip().split('Scenario: ')
scenarios = [s for s in scenarios if s.strip()]  # Remove empty strings if any

# Function to parse each scenario into a dictionary
def parse_scenario(scenario_text):
    lines = scenario_text.strip().split('\n')
    scenario_name = lines[0].strip()
    test_case = {}
    for line in lines[1:]:
        if line.startswith('testSteps:'):
            test_case['testSteps'] = []
            continue
        if line.startswith('  - Step '):
            test_case['testSteps'].append(line.strip())
            continue
        if ':' in line:
            key, value = line.split(':', 1)
            test_case[key.strip()] = value.strip()
    return scenario_name, test_case

# Parse all scenarios
parsed_scenarios = {}
for scenario in scenarios:
    name, details = parse_scenario(scenario)
    parsed_scenarios[name] = details


filename = os.path.join(path,directory_name) + '.json'
with open(filename, 'w') as f:
    json.dump(parsed_scenarios, f, indent=4)

# print(parsed_scenarios)
print(parsed_scenarios["Lesson Creation with Duplicate Name-Raw1.txt"]["testSteps"])
# # Write each scenario to a separate JSON file
# for scenario_name, scenario_details in parsed_scenarios.items():
#     filename = scenario_name + '.json'
#     with open(filename, 'w') as f:
#         json.dump(scenario_details, f, indent=4)

# print("JSON files created successfully.")
```
