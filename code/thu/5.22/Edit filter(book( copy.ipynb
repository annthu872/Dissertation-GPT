```python
from openai import OpenAI
import json
import os

GPT_MODEL_4 = "gpt-4-0125-preview"
OPEN_API_KEY = os.getenv("OPENAI_API_KEY")

client = OpenAI()
model = GPT_MODEL_4

def ask(prompt, client, model, temperature = 0):
    response = None
    
    response = client.chat.completions.create(
      model=model,
      messages=prompt,
      temperature=temperature,
    )

    return response.choices[0].message.content

def askJSON(prompt, client, model, temperature = 0):
    response = None
    
    response = client.chat.completions.create(
      model=model,
      messages=prompt,
      temperature=temperature,
      response_format={ "type": "json_object" },
    )

    return response.choices[0].message.content
```

```python
FILTER_SYSTEM_PROMPT="""
You will be provide with a use case and a list of test scenario.
Based on information in the use case flow, define what test scenario is necessary to test the use case and remove duplicate test scenarios.
Remove test case related with system error that not be mentioned in use case such as: Page Load Failure, System Error, Network Error,Non-Existent,Insufficient Information, Incomplete Information ... (and use case not mentioned these flow)
Make sure test scenarios filted cover all the flow of use case and every scenarios in the response is unique.
Return scenario name only.
"""
# Make sure every scenarios in the response is unique.
```

```python
print("Display the list of authors")
filtercontent = """
Use case: 
Main scenario:
User selects item "Authors";
Application displays the list of the authors (up to 10 items on each page);
To see the next 10 records of the authors in the database, the user needs to click the "Next" button (below the records in the pagination block) or the following number of the page the user wants to see.

The list displays the following columns:
id - unique author number
Name
Bio - author's biography

Test scenarios:
Flow scenario:
1. Display First Page of Authors
2. Navigate to Next Page of Authors
3. Navigate to a Specific Page of Authors
4. Last Page Display and Navigation Behavior
5. Navigation with Insufficient Authors for Further Pagination
"""
promptFilter = [
    { "role": "system", "content": FILTER_SYSTEM_PROMPT},
    { "role": "user", "content": filtercontent}
  ]
filter_gpt_response = ask(promptFilter, client, model)
print("\n" + "Final result:" + filter_gpt_response)
```
