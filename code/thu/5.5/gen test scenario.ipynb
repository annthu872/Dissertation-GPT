```python
from openai import OpenAI
import json
import os

GPT_MODEL_3_5 = "gpt-3.5-turbo-1106"
GPT_MODEL_4 = "gpt-4-0125-preview"

OPEN_API_KEY = os.getenv("OPENAI_API_KEY")

client = OpenAI()
model = GPT_MODEL_4

def ask(prompt, client, model, temperature = 0):
    response = None
    
    response = client.chat.completions.create(
      model=model,
      messages=prompt,
      temperature=temperature,
      # response_format={ "type": "json_object" },
    )

    return (response.choices[0].message.content)
```

```python
def read_file_into_string(file_path):
    encodings = ['utf-8', 'latin-1', 'utf-16', 'cp1252']
    
    for encoding in encodings:
        try:
            with open(file_path, 'r', encoding=encoding) as file:
                file_content = file.read()
            return file_content
        except UnicodeDecodeError:
            print(f"Failed to decode using {encoding} encoding.")
            continue
        except FileNotFoundError:
            print(f"File '{file_path}' not found.")
            return None
        except Exception as e:
            print(f"Error reading file: {e}")
            return None
    
    print("All tried encodings failed. Unable to read the file.")
    return None
```

```python
file_path = r"D:\GPT-testing\SpecificationData\MatchaEnglishWebsite\AddQuestion .txt"
scenario = """Admin navigates to lesson management page from homepage """
print(file_path)
```

```python
# Example usage:
# file_path = "D:\GPT-testing\SpecificationData\Adventura - Flow only\Adventura - Filter.txt"
input = "Test scenario: "+scenario+"\nUse case: "+ read_file_into_string(file_path)
if input is not None:
    print(input)
```

```python
SYSTEM_PROMPT_gen_test_scenario ="""
I want you to act as a software tester.
Your task is to read the test scenario's name and the corresponding use case specification to predict the goal of the test scenario.
Also, generate a test case includes test steps based on those information to cover the test scenario.
Rules for generating test case:
- If that test case needs login as a specific role to proceed, do not put login at step 1. Instead, put the login as that specific role in the pre condition.
- If in the use case specification do not describe how to go to the page to do the test case, do not put it in the step. Instead, tell the tester that they need to be in that page in the pre condition.
"""
```

```python
promptExtractCondtition = [
    { "role": "system", "content": SYSTEM_PROMPT_gen_test_scenario},
    { "role": "user", "content": input}
]

gpt_response = ask(promptExtractCondtition, client, model)
print(gpt_response)
```
