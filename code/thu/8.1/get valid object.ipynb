```python
import pandas as pd

# Load the Excel file
file_path = r"D:\Matcha.xlsx"  # Update this with your actual file path
# df = pd.read_excel(file_path, sheet_name='Trang tinh1')
df = pd.read_excel(file_path,header=None)

# Iterate through the DataFrame to keep only the valid scenarios
def filter_scenarios(df):
    valid_scenarios = []
    use_case = None
    
    for index, row in df.iterrows():
        if pd.isna(row[0]):
            use_case = None
        elif row[0].endswith('.txt'):
            use_case = row[0]
        elif row[1] == 'x' and pd.isna(row[3]):
            valid_scenarios.append([use_case, row[0]])
    
    return valid_scenarios

filtered_scenarios = filter_scenarios(df)

# Create a new DataFrame with the filtered scenarios
filtered_df = pd.DataFrame(filtered_scenarios, columns=['Use Case', 'Scenario'])

# Save the filtered DataFrame to a new Excel file
filtered_file_path = 'Matcha.xlsx'
filtered_df.to_excel(filtered_file_path, index=False)

print(f"Filtered scenarios saved to {filtered_file_path}")
```
